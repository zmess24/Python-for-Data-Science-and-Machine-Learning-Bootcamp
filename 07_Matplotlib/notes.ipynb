# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

## 1. Functional Method

```python
plt.plot(x,y)
plt.xlabel('X Label') # Add x label
plt.ylabel('Y Label') # Add y label
plt.title('Title') # Add title
```

```python
# Multiple plots on same canvas
plt.subplot(1,2,1)
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(y, x, 'b')
```

## 2. Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('y Label')
axes.set_title('Title Goes Here!')
```

```python
fig = plt.figure()
# Left, Bottom, Width, Height
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
# axes2 = fig.add_axes([0.4, 0.15, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(y, x)

axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

## 3. Subplots

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

## 4. Figure Size & DPI

```python
fig = plt.figure(figsize=(8,2)) # X Inches by Y inches

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2)) # X Inches by Y inches

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# SAVE A FIGURE
fig.savefig('my_picture.png', dpi=200)
```

```python
# ADD LEGENDS
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.set_title("Title")
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")

ax.legend(loc=0) # Default location is top right corner
```

## 5. Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# Can accept hex colors, alpha = transparency
ax.plot(x, y, 
        color="orange", 
        linewidth=3,
        alpha=0.5, 
        linestyle='--', 
        marker='o',
        markersize=10,
        markerfacecolor='red',
        markeredgewidth=3,
        markeredgecolor='green'
        )
```

## 6. Axis Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color="purple", linewidth=3, linestyle='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
