```python
import chart_studio.plotly as py
import pandas as pd
import numpy as np
import cufflinks as cf

from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df1 = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
df1.head()
```

```python
df2 = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [32,43,50]})
df2
```

```python
df1.iplot()
```

```python
df1.iplot(kind='scatter', x='A', y='B', mode='markers', size=20)
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df1.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y': [10,20,30,20,10], 'z': [5, 4, 3, 2, 1]})
```

```python
df3.iplot(kind='surface')
```

```python
df1['A'].iplot(kind='hist', bins=50)
```

```python
df1[['A', 'B']].iplot(kind='spread')
```

```python
df1.iplot(kind='bubble', x='A', y='B', size='C')
```
