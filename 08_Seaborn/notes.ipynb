# Seaborn

```python
import seaborn as sns
```

```python
%matplotlib inline
```

## 1. Distribution Plots

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.displot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="kde")
```

```python
sns.pairplot(tips, hue="sex", palette="coolwarm")
```

```python
sns.rugplot(tips["total_bill"])
```

## 2. Categorical Plots

```python
import numpy as np

sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.std)
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x='day', y="total_bill", data=tips, huee="sex", split=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, hue="sex")
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color="black")
```

```python
sns.factorplot(x="day", y="total_bill", data=tips, kind="bar")
```

## 3. Matrix Plots

```python
flights = sns.load_dataset('flights')

flights.head()
```

```python
tc = tips.corr(numeric_only=True)
```

```python
sns.heatmap(tc, annot=True, cmap="coolwarm")
```

```python
fp = flights.pivot_table(index="month", columns="year", values="passengers")
```

```python
sns.heatmap(fp, cmap="magma")
```

```python
sns.heatmap(fp, cmap="magma", linecolor="black", linewidths=1)
```

```python
sns.clustermap(fp, cmap="coolwarm", standard_scale=1)
```

## 4. Grids

```python
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
sns.pairplot(data=iris)
```

```python
g = sns.PairGrid(data=iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(data=iris)

g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, col="time", row="smoker")
g.map(sns.histplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col="time", row="smoker")
g.map(plt.scatter, 'total_bill', 'tip')
```

## 5. Regression Plots

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", markers=["o", "v"],
          scatter_kws={'s': 100})
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col="sex")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col="sex", row="time")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col="day", hue="sex",
          aspect=0.6)
```

## 6. Style & Color

```python
plt.figure(figsize=(12,5))
sns.set_style('whitegrid')
sns.countplot(x="sex", data=tips)
sns.despine(left=True, bottom=False)
```
